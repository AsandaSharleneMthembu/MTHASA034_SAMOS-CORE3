# Part 2 — CTD Profiles: Temperature and Salinity

Reads the CSV file from Part 1 and plots temperature and salinity as vertical profiles,
displayed as small multiples with a shared y-axis.

**Input:** Change `INPUT_FILE` to your CSV filename and `OUTPUT_FILE` to your desired figure name.

```python
import os
print(os.listdir())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ── User inputs ────────────────────────────────────────────────────────────────
INPUT_FILE  = '11_29_2008_CTD.csv'
OUTPUT_FILE = 'CTD_profiles.png'
# ──────────────────────────────────────────────────────────────────────────────

def ddmm2dd(ddmm):
    # Converts position from degrees-minutes (DDMM.cccc) to decimal degrees
    # e.g. 4530.5 → 45.508 (45° 30.5') — sign (N/S, E/W) handled separately
    thedeg = np.floor(ddmm / 100.)
    themin = (ddmm - thedeg * 100.) / 60.
    return thedeg + themin

# Load the CSV produced in Part 1
df = pd.read_csv(INPUT_FILE)

# If dataset contains lat/lon in DDMM format, convert to decimal degrees
# df['latitude']  = df['lat_ddmm'].apply(ddmm2dd)
# df['longitude'] = df['lon_ddmm'].apply(ddmm2dd)

# Create two side-by-side panels sharing the same y-axis (depth)
# sharey=True ensures both profiles are directly comparable vertically
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(8, 7))

# --- Temperature panel ---
# Plot temperature vs depth as a continuous line (no markers)
ax1.plot(df['temperature'], df['depth'], color='tomato', linewidth=1.5)
ax1.set_xlabel('Temperature (°C)', fontsize=12)
ax1.set_ylabel('Depth (m)', fontsize=12)
ax1.set_title('Temperature', fontsize=13, fontweight='bold')
ax1.invert_yaxis()                    # depth increases downward (oceanographic convention)
ax1.grid(True, linestyle='--', alpha=0.5)

# --- Salinity panel ---
# sharey=True means the inverted y-axis from ax1 is automatically inherited here
ax2.plot(df['salinity'], df['depth'], color='steelblue', linewidth=1.5)
ax2.set_xlabel('Salinity (PSU)', fontsize=12)
ax2.set_title('Salinity', fontsize=13, fontweight='bold')
ax2.grid(True, linestyle='--', alpha=0.5)

# Overall figure title and layout
fig.suptitle('CTD Profiles', fontsize=15, fontweight='bold')
plt.tight_layout()                    # prevent overlapping labels

plt.savefig(OUTPUT_FILE, dpi=150, bbox_inches='tight')
plt.show()
```

## Part 3: AI-Assisted Reflection Prompt

Explain why datetime indexing is important in oceanographic time series analysis and discuss the advantages of using descriptive statistics and visualisation techniques to interpret environmental variability during a research cruise.

```python


# ============================================================
# PART 3 — Oceanographic Time Series Analysis
# Dataset: SAA2_WC_2017_metocean_10min_avg.csv
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ── User inputs ────────────────────────────────────────────────────────────────
INPUT_FILE     = 'SAA2_WC_2017_metocean_10min_avg.csv'
OUTPUT_SCATTER = 'wind_airtemp_scatter.png'
# ──────────────────────────────────────────────────────────────────────────────


# ============================================================
# 1. LOAD DATA
# TIME_SERVER is used as the datetime index for time-based slicing
# ============================================================

df = pd.read_csv(
    INPUT_FILE,
    parse_dates=['TIME_SERVER'],    # parse TIME_SERVER as datetime
    index_col='TIME_SERVER',        # set as index for time-based slicing
    na_values=['NaN', 'nan', -9999, 'N/A', '']
)

print(f'Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns')
print(f'Time range: {df.index[0]} → {df.index[-1]}')
print(f'\nMissing values per column:\n{df.isnull().sum()}')
df.head()


# ============================================================
# 2. SUBSET — Start of cruise to southernmost point (July 4)
# ============================================================

df_cruise = df.loc[:'2017-07-04']

print(f'Cruise subset: {df_cruise.shape[0]} rows')
print(f'Period: {df_cruise.index[0]} → {df_cruise.index[-1]}')


# ============================================================
# 3. TEMPERATURE TIME SERIES
# TSG_TEMP is the thermosalinograph sea surface temperature
# Grayscale style chosen for print-friendly academic output
# ============================================================

plt.style.use('grayscale')

fig, ax = plt.subplots(figsize=(12, 4))

ax.plot(df_cruise.index, df_cruise['TSG_TEMP'], linewidth=0.8)

ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Temperature (°C)', fontsize=12)
ax.set_title('Sea Surface Temperature — Cruise Period (to July 4)', fontsize=13, fontweight='bold')
ax.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('temperature_timeseries.png', dpi=150, bbox_inches='tight')
plt.show()


# ============================================================
# 4. SALINITY HISTOGRAM
# TSG_SALINITY from the thermosalinograph
# Bins set from 30–35 PSU at 0.5 intervals
# ============================================================

plt.style.use('default')

fig, ax = plt.subplots(figsize=(7, 5))

bins = np.arange(30, 35.5, 0.5)

ax.hist(
    df_cruise['TSG_SALINITY'].dropna(),
    bins=bins,
    color='steelblue',
    edgecolor='white',
    linewidth=0.6
)

ax.set_xlabel('Salinity (PSU)', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
ax.set_title('Salinity Distribution — SA Aguhlhas', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.4, axis='y')
plt.tight_layout()
plt.savefig('salinity_histogram.png', dpi=150, bbox_inches='tight')
plt.show()


# ============================================================
# 5. DESCRIPTIVE STATISTICS
# IQR is robust to outliers — important for shipboard sensor data
# ============================================================

def compute_stats(series):
    """Return mean, std, and IQR for a pandas Series."""
    return pd.Series({
        'Mean' : series.mean(),
        'Std'  : series.std(),
        'IQR'  : series.quantile(0.75) - series.quantile(0.25)
    })

stats = pd.DataFrame({
    'Temperature (°C)' : compute_stats(df_cruise['TSG_TEMP']),
    'Salinity (PSU)'   : compute_stats(df_cruise['TSG_SALINITY'])
}).T.round(3)

print('\nDescriptive Statistics:')
display(stats)


# ============================================================
# 6. SCATTER PLOT — Wind Speed vs Air Temperature
# WIND_SPEED_TRUE and AIR_TEMPERATURE plotted against each other
# LATITUDE encoded as colour to reveal spatial structure
# along the cruise track
# ============================================================

fig, ax = plt.subplots(figsize=(8, 6))

sc = ax.scatter(
    df_cruise['WIND_SPEED_TRUE'],
    df_cruise['AIR_TEMPERATURE'],
    c=df_cruise['LATITUDE'],
    cmap='viridis',
    s=5,
    alpha=0.7
)

cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('Latitude (°)', fontsize=11)

ax.set_xlabel('True Wind Speed (m/s)', fontsize=12)
ax.set_ylabel('Air Temperature (°C)', fontsize=12)
ax.set_title('Wind Speed vs Air Temperature\ncoloured by Latitude', fontsize=13, fontweight='bold')
ax.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()

plt.savefig(OUTPUT_SCATTER, dpi=300, bbox_inches='tight')
print(f'Scatter plot saved as "{OUTPUT_SCATTER}"')
plt.show()
```
